```python
# 입사 성적(list, for, if)을 처리하는 프로그램을 제작하시오.
# 성적을 list에 저장 할 것.
# python: 90, java: 80, html: 100, dbms: 70, tensorflow: 100
# 총점, 평균, 등급(90 ~ 100: A, 80 ~ 89: B, 70 ~ 79: C, 60 ~ 69: D, 그외는 F를 출력할 것)

# 총점: 440, 평균: 88, 등급: B

# Hint: list, for, elif

list1 = [90, 80, 100, 70, 100]
sum = 0
for i in range(0, 5):
  sum += list1[i]
print(f"총점: {sum},", end ='')
average = int(sum / 5)
print(f" 평균:{average},", end ='')
if 90 <= average <= 100:
  print(" 등급 : A")
elif 80 <= average < 90:
  print(" 등급 : B")
elif 70 <= average < 80:
  print(" 등급 : C")
elif 60 <= average < 70:
  print(" 등급 : D")
else:
  print(" 등급 : F")
```

```python
# 2차원
sungjuk = [['가길동', 100, 90, 80],
           ['나길순', 90, 95, 85],
           ['다길순', 70, 75, 90]]

for row in sungjuk:
  print(row)
  name = row[0]
  tot = row[1] + row[2] + row[3]
  avg = tot / 3
  print(f'성명: {name}, 총점: {tot}, 평균: {avg:1f}')
```

```python
# 각 영업부서의 판매실적을 10만원당 '*'하나로 대응하여 출력하시오.
# 출력
# 95 :***********************************************************************************************
# 80 :********************************************************************************
# 75 :***************************************************************************
# 71 :***********************************************************************
# 67 :*******************************************************************
# 50 :**************************************************
# 45 :*********************************************
# 필수 구현
sales = [950, 750, 500, 450, 800, 670, 710]

for i in range(len(sales)): # 0 ~ 6
    sales[i] = int(sales[i] / 10) # int : 반올림 안됨

# print(sales)
sales.sort(reverse=True)
# print(sales)

for i in range(len(sales)): # 0 ~ 6
    print(sales[i], ':', '*' * sales[i])
```

```python
# 데이터 분할
import random

random.seed(1)
data = [random.randint(1, 100) for i in range(100)]
print(data)
```
