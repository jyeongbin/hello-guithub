```python
# 신체질량지수(BMI) = (몸무게(kg) / 신장의 제곱) * 10000
# 몸무게(kg) 입력:64
# 키(신장, cm) 입력:167
# BMI: 22.948115744558788

kg = int(input('(kg) 입력: '))
cm = int(input('키(신장, cm) 입력: '))
BMI = kg / (cm ** 2) * 10000
print(f'BMI: {BMI}')
```

```python
# round 함수 : 반올림
price = 10555.56755
print(round(price, 4)) # 5 에서는 반올림이 안됨
price = 10555.56755001
print(round(price, 4)) # 5 초과시 반올림 발생
print(round(price, 3))
print(round(price, 2))
print(round(price, 1))
print(round(price, 0))
print(round(price, -1))
print(round(price, -2))
```

```python
# 지폐수 처리
# 돈의 액수를 적어주십시오. :79400
# 50000원권 : 1
# 10000원권 : 2
# 5000원권 : 1
# 1000원권 : 4

price = int(input("돈의 액수를 적어주십시오.:"))
오만원 =  price // 50000
print(f"50000원권 : {오만원}")
total = price - (50000 * 오만원)

만원 = total // 10000
print(f"10000원권 : {만원}")
total = total - (10000 * 만원)

오천원 = total // 5000
print(f"5000원권 : {오천원}")
total = total - (5000 * 오천원)

천원 = total // 1000
print(f"1000원권 : {천원}")
```

```python
# 토스뱅크의 보통예금은 년 2% 이자를 복리로 계산해준다. 금액을 입력했을 때 매일 받는 이자는?
# - 이자소득의 원천징수세율은 14%를 적용 할 것.

# 금액 입력:50000000
# 원금: 50,000,000 원
# 년이자: 1,000,000 원
# 월이자: 83,333 원
# 오늘이자: 2,739 원
# 세후이자: 2,355 원

price = int(input("금액 입력:"))
ija = int(price * 0.02)
day = int(ija/365)
receive = int(day-(day*0.14))
print(f"원금 : {format(price, ',')} 원")
print(f"년이자: {format(price // 50, ',')} 원")
print(f"월이자:  {format((price //50) // 12 , ',')} 원")
print(f"오늘이자: {format(day, ',')} 원")
print(f"세후이자: {format(receive, ',')} 원")
```
