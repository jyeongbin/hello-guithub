```python
# 1 ~ 1000까지의 수 중에서 2 and 3 and 4 and 5의 배수를 출력하시오.
# hint: for, if
# 60  120  180  240  300  360  420  480  540  600  660  720  780  840  900  960
i = 60
for i in range(i, 1000):
  if(i%2 == 0) and (i%3 == 0) and (i%4 == 0) and (i%5 == 0):
    print(i, end=' ')
```

```python
# 1 ~ 1000까지의 수 중에서 2 and 3 and 4 and 5의 배수와 갯수를 출력하시오.
# hint: for, if
# 60  120  180  240  300  360  420  480  540  600  660  720  780  840  900  960
# cnt: 16

i = 60
cnt = 0
for i in range(i, 1001):
  if(i%2 == 0) and (i%3 == 0) and (i%4 == 0) and (i%5 == 0):
    print(i, end=' ')
    cnt+=1 # 자기자신의 메모리를 계속 덮어씀
print() #줄 바꿈
print('cnt:', cnt)
```

```python
# 특정 단어를 입력 받아 순환 횟수 만큼 출력 할 것.
# 입력: 단어, 순환 횟수
# 단어: ★
# 순환 횟수: 3
# ★★★
# print('-' * 30)

# 특정 단어를 입력 받아 순환 횟수만큼 출력할 것

point = int(input("순환 횟수: "))
for i in range(1 ,point+1):
    print('★', end= '')

    if i >= 10:   # 출력을 최대 10회로 제한
      break;      # 인접한 반복문을 탈출
```

```python
# 입력 받은 숫자가 짝수인지 홀수인지를 출력하고 9999를 입력하면 프로그램을 종료하시오.
# - 입력: 정수 1개
# 힌트: for문 종료(탈출) 선언 필요, break
# for문 순환 횟수는 100이상 자유롭게 지정: 입력받을 필요없음

# 입력: 3
# 홀수
# 입력: 2
# 짝수
# 입력: 9999

for i in range(100):
    su = int(input('입력: '))
    if su == 9999:
      break
    else:
      if su %2 == 0:
        print("짝수")
      else:
        print("홀수")
```

```python
# 입력 받은 숫자가 짝수인지 홀수인지 여부와 갯수를 출력하고 9999를 입력하면 프로그램을 종료하시오.
# - 입력: 정수 1개
# - 출력

# 입력: 1
# 홀수입니다.
# 짝수 갯수 :  0 홀수 갯수 :  1
# ------------------------------
# 입력: 2
# 짝수입니다.
# 짝수 갯수 :  1 홀수 갯수 :  1
# ------------------------------
# 입력: 3
# 홀수입니다.
# 짝수 갯수 :  1 홀수 갯수 :  2
# ------------------------------
# 입력: 9999
# 사용해 주셔서 감사합니다.

even = 0
odd = 0
for i in range(1,10000):
  su = int(input("입력 : "))
  if(su == 9999):
    print("사용해 주셔서 감사합니다.")
    break # for 종료
  else:
    if su % 2 == 0:
      print("짝수입니다.")
      even = even + 1
      print('-' * 30)
    else:
      print("홀수입니다.")
      odd = odd + 1
      print('-' * 30)
  print(f"짝수 갯수:  {even} 홀수 갯수 : {odd}")
```

```python
# 3번 순환하면서 숫자를 입력 받아, 입력 받은 숫자까지의 합을 출력 할 것.
# - 입력: 정수 1개
# - 출력:
#  입력: 1
#  합계: 1
#  입력: 3
#  합계: 4
#  입력: 5
#  합계: 9

sum = 0
for i in range(3):
  su = int(input("입력: "))
  sum = sum + su
  print(f"합계: {sum}")
```
